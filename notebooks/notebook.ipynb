## Data Processing Stuff....

```python
df = df.drop(columns=["Patient_ID"])
```

```python
X = df.drop(columns=["Survival_Prediction"])
y = df["Survival_Prediction"]
```

```python
categorical_columns = X.select_dtypes(include=["object"]).columns
label_encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] =le

    


```

```python
X.head()
```

```python
X.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

## Feature Selction. CHI-Square Testing....

```python
X_cat = X_train.select_dtypes(include=["int64","float64"])
chi2_selector = SelectKBest(score_func=chi2, k="all")
chi2_selector.fit(X_cat,y_train)
```

```python
chi2_scores= pd.DataFrame({
    'Feature': X_cat.columns,
    'Chi2_Score': chi2_selector.scores_
}).sort_values(by='Chi2_Score', ascending=False)
```

```python
chi2_scores
```

```python
top_features = chi2_scores.head(5)["Feature"].tolist()
```

```python
X = X[top_features]
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=42, stratify=y)
```

```python
X_train.shape
```

## MODEL BUILDING....

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


```

```python
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
```

```python
gb.fit(X_train,y_train)
```

```python
y_pred = gb.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
report = classification_report(y_test, y_pred)
```

```python
print(report)
```

```python
y_proba = gb.predict_proba(X_test)[:,1] if len(y.unique()) == 2 else None
```

```python
roc_auc_score(y_test, y_proba)
```

```python
plt.figure(figsize=(5,5))
plt.barh(top_features, gb.feature_importances_, color="skyblue")
plt.gca().invert_yaxis()
```
